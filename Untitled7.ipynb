```python
import pandas as pd

climate_df = pd.read_csv(
    "/content/Environment_Temperature_change_E_All_Data_NOFLAG.csv",
    encoding="latin1"
)

print(climate_df.columns)
```

```python
# Keep only temperature change rows
climate_df = climate_df[climate_df["Element"] == "Temperature change"]

# Select year columns
year_cols = [col for col in climate_df.columns if col.startswith("Y")]

climate_df = climate_df[["Area", "Months"] + year_cols]
```

```python
climate_df["Avg_Temp_Change"] = climate_df[year_cols].mean(axis=1)
```

```python
season_map = {
    "December": "Winter", "January": "Winter", "February": "Winter",
    "March": "Summer", "April": "Summer", "May": "Summer",
    "June": "Rainy", "July": "Rainy", "August": "Rainy",
    "September": "Spring", "October": "Spring", "November": "Spring"
}

# Keep only real months
climate_df = climate_df[climate_df["Months"].isin(season_map.keys())]

# Map months to seasons
climate_df["Season"] = climate_df["Months"].map(season_map)
```

```python
climate_season = (
    climate_df
    .groupby("Season", as_index=False)["Avg_Temp_Change"]
    .mean()
)
```

```python
climate_season.to_csv("climate_seasonal_temperature.csv", index=False)
```

```python
# Save the processed climate dataset
climate_season.to_csv("climate_seasonal_temperature.csv", index=False)

print("Saved as climate_seasonal_temperature.csv")
```
